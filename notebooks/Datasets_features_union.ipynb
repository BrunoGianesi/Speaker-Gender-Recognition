```python
import pandas as pd
import os
```

# Union Concatenating

```python
dataset = 'MLS'
project_root = os.path.dirname(os.getcwd())
```

```python
Features = pd.read_csv(f'{project_root}/data/{dataset}/Features_data.csv')
F0 = pd.read_csv(f'{project_root}/data/{dataset}/F0_data.csv')
MFCCs = pd.read_csv(f'{project_root}/data/{dataset}/MFCCs_data.csv')
```

```python
result = pd.concat([Features, F0, MFCCs], axis=1)
result = result.loc[:,~result.columns.duplicated()]
result
```

```python
result.to_csv(f'{project_root}/data/{dataset}/Features_F0_MFCCs_data.csv', index=False)
```

# Union joining

```python
dataset = 'MLS_split'
project_root = os.path.dirname(os.getcwd())
```

```python
Features = pd.read_csv(f'{project_root}/data/{dataset}/Features_data.csv')
F0 = pd.read_csv(f'{project_root}/data/{dataset}/F0_data.csv')
MFCCs = pd.read_csv(f'{project_root}/data/{dataset}/MFCCs_data.csv')
```

```python
final = (pd.merge(Features,F0,on=['FileName', 'Gender']))
final = pd.merge(final ,MFCCs,on=['FileName', 'Gender'])
final
```

```python
final.to_csv(f'{project_root}/data/{dataset}/Features_F0_MFCCs_data.csv', index=False)
```
