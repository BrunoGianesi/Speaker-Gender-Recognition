```python
import os
import shutil
```

## Create dataset only with the audio files

```python
project_root =  os.path.dirname(os.path.dirname(os.getcwd()))
source = os.path.join(project_root, "CETUC", "alcaim")
destination = os.path.join(project_root, "CETUC", "Full")
# protect so you dont run this accidentally
are_you_sure = True
if are_you_sure:
    folders = os.listdir(source)
    for i, folder in enumerate(folders):
        path = os.path.join(source, folder)
        if os.path.isdir(path):
            files = os.listdir(path)
            for file in files:
                file_path = os.path.join(path, file)    
                if file.endswith('.wav'):
                    shutil.copy(file_path,destination)
        print(f"{(i/len(folders))*100}%\r")
```
